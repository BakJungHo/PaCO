========

```python
import pandas as pd
```

```python
uriage_df = pd.read_csv('/content/drive/MyDrive/특강 문제 풀이 1회차/num_1/uriage.csv') #pd.read_csv("./num_1/uriage.csv")
uriage_df 
```

2번 문제 
===========

1. num_2 폴더 내에 있는 csv 파일을 모두 데이터프레임으로 불러온다. 
2. transaction_1과 transaction_2를 유니언 결합을 한다. 
3. transaction_detail_1과 transaction_detail_2를 유니언 결합을 한다. 
4. 2,3 번의 데이터프레임의 값을 조인 결합을 한다. 
5. 4번 작업한 데이터프레임을 customer_master, item_master 와 조인 결합한다. 
6. quantity와 item_price의 곱한 값으로 price라는 컬럼을 생성한다. 
7. payment_date컬럼의 데이터형태를 datetime으로 변경한다. 
8. payment_date값의 년 월값만 추출하여 payment_month 컬럼을 생성한다. 
9. groupby를 이용하여 월별, 상품별 price와 quantity의 합계를 출력한다. 

```python
import pandas as pd
```

```python
customer_master = pd.read_csv('/content/drive/MyDrive/특강 문제 풀이 1회차/num_2/customer_master.csv')
customer_master
```

```python
item_master = pd.read_csv('/content/drive/MyDrive/특강 문제 풀이 1회차/num_2/item_master.csv')
item_master
```

```python
transaction_1 = pd.read_csv('/content/drive/MyDrive/특강 문제 풀이 1회차/num_2/transaction_1.csv')
transaction_1
```

```python
transaction_2 = pd.read_csv('/content/drive/MyDrive/특강 문제 풀이 1회차/num_2/transaction_2.csv')
transaction_2
```

```python
transaction_detail_1 = pd.read_csv('/content/drive/MyDrive/특강 문제 풀이 1회차/num_2/transaction_detail_1.csv')
transaction_detail_1
```

```python
transaction_detail_2 = pd.read_csv('/content/drive/MyDrive/특강 문제 풀이 1회차/num_2/transaction_detail_2.csv')
transaction_detail_2
```

```python
transaction_df = pd.concat([transaction_1, transaction_2], ignore_index=True) # 유니온 결합: 데이터 프레임의 행을 추가하는 작업 - concat함수를 사용한다. ignore_index = True 는 인덱스를 재배열 하는 것.
transaction_df
```

```python
transaction_detail_df = pd.concat([transaction_detail_1, transaction_detail_2], ignore_index=True)
transaction_detail_df
```

```python
join_data = pd.merge(transaction_detail_df, transaction_df[['transaction_id','payment_date','customer_id']], on = "transaction_id", how="inner")
join_data
```

```python
print(len(transaction_detail_df))
print(len(transaction_df))
print(len(join_data))

# 데이터프레임의 행의 수가 증가한 것을 확인할 수 있다. 
```

```python
join_data = pd.merge(join_data, customer_master, on = 'customer_id', how='left')
join_data = pd.merge(join_data, item_master, on = "item_id", how='left')
join_data
```

```python
join_data['price'] = join_data['quantity'] * join_data['item_price']
join_data[['quantity','item_price','price']]
```

```python
join_data['payment_date'] = pd.to_datetime(join_data['payment_date'])
join_data['payment_date']
```

```python
join_data['payment_month'] = join_data['payment_date'].dt.strftime('%Y%m')
join_data['payment_month']
```

```python
# groupby를 이용하여 월별, 상품별 price와 quantity의 합계를 출력한다.

join_data.groupby(['payment_month','item_name']).sum()[['price','quantity']]
```

3번 문제
==========

1. num_3 폴더내의 drink 파일을 데이터프레임으로 불러온다. 
2. 대륙 컬럼인 continent 결측 데이터를 OT로 지정한다. 
3. 대륙별 spirit_servings의 평균, 최소, 최대, 합계를 계산한다. 
4. 전체 평균보다 많은 알코올 섭취하는 대륙을 구한다. 
5. 3번 과정의 데이터를 바그래프를 이용하여 시각화 한다. 

```python
drink_df = pd.read_csv('/content/drive/MyDrive/특강 문제 풀이 1회차/num_3/drinks.csv')
drink_df
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
drink_df.isnull().sum()
```

```python
drink_df = drink_df.replace(to_replace = np.nan, value = "OT")
#결측 데이터를 처리한다 : drink_df['continent'] = drink_df['continent'].fillna('OT')
```

```python
drink_df.isnull().sum()
```

```python
# 대륙별 spirit_servings의 평균, 최소, 최대, 합계를 계산한다
# 모든 열에 여러 함수를 매핑 : group객체.agg(함수1, 함수2, 함수3 ...)
# 각 열마다 다른 함수를 매핑 : group객체.agg({'열1':함수1, '열2':함수2, ...})

drink_df.groupby(['continent']).sum()['spirit_servings']
```

```python
result = drink_df.groupby('continent').spirit_servings.agg(['mean', 'min', 'max', 'sum'])
result
```

```python
drink_df.groupby('continent').spirit_servings.describe()
```

```python
# 전체 평균보다 많은 알코올 섭취하는 대륙을 구한다.

total_mean = drink_df.total_litres_of_pure_alcohol.mean()
continent_mean = drink_df.groupby('continent')['total_litres_of_pure_alcohol'].mean()
continent_over_mean = continent_mean[continent_mean >= total_mean]
print(continent_over_mean)
```

```python
# 대륙별 spirit_servings의 평균, 최소, 최대, 합계를 시각화합니다.
n_groups = len(result.index)
means = result['mean'].tolist()
mins = means = result['min'].tolist()
maxs = result['max'].tolist()
sums = result['sum'].tolist()

index = np.arange(n_groups)
bar_width = 0.1

rects1 = plt.bar(index, means, bar_width,
                 color='r',
                 label='Mean')

rects2 = plt.bar(index + bar_width, mins, bar_width,
                 color='g',
                 label = 'Min')

rects3 = plt.bar(index +bar_width * 2, maxs, bar_width,
                 color = 'b',
                 label = 'Max')

rects4 = plt.bar(index + bar_width * 3, sums, bar_width,
                 color = 'y',
                 label = 'Sum')

plt.xticks(index, result.index.tolist())
plt.legend()
plt.show()
```
