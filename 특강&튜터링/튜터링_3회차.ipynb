```python
titanic_df.dropna()

# 결측치 제거
# 결측치를 포함하는 모든 행 제거 -> axis = 0이 default
```

```python
titanic_df.dropna(subset = ['Age'])

# 특정 열의 결측치만 제거
```

```python
titanic_df.dropna(subset=['Cabin'])
```

```python
nan = titanic_df[titanic_df['Age'].isna() == True]['Age']

# 결측치가 있는 행만 추출한 DataFrame에서 다시 'Age' 컬럼만 추출

nan
```

```python
nan.dtypes
```

```python
nan = pd.DataFrame(nan)
nan 
```

```python
nan.reset_index(drop = False, inplace=True)
```

```python
nan.drop(['level_0'], axis=1, inplace=True)
```

```python
nan.drop(['index'], axis=1, inplace=True)
```

```python
nan
```

```python
fill_1 = nan.fillna(0)

#결측치를 0으로 채우기 
fill_1
```

```python
nan.iloc[0,:]= 0
nan.iloc[-1, :] = 176

nan
```

```python
fill_2 = nan.interpolate()

# 두 수치 차이를 채워주기 위한 함수
# 보간법의 종류는 다양
# default는 'linear', 즉 선형성을 띠도록 채워줌

fill_2
```

```python
nan.interpolate('nearest')

# 가까운 값으로 채워줌
```

```python
# 'Cabin' 칼럼의 결측치를 'C000'으로 채우기
# 'Age' 칼럼의 결측치를 전체 나이의 평균으로 채우기 
```

```python
titanic_df['Cabin'] = titanic_df['Cabin'].fillna('C000')
titanic_df.isnull().sum()
```

```python
titanic_df['Age'].mean()
```

```python
titanic_df['Age'] = titanic_df['Age'].fillna(titanic_df['Age'].mean())
titanic_df.isnull().sum()
```
