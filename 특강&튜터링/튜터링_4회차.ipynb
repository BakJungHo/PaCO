```python
# 계산기
# calculator()라는 함수를 사용하면, 숫자를 2개 넣고 결과값 도출

def plus(x, y):
  return x + y

def minus(x, y):
  return x - y

def mul(x, y):
  return x * y

def dev(x, y):
  return x / y

def calculator():
  num1 = int(input("num1: "))
  num2 = int(input("num2: "))
  calculate = input("연산자를 입력해주세요.(+, -, *, /)")

  if calculate == "+":
    result = plus(num1, num2)
  elif calculate == "-":
    result = minus(num1, num2)
  elif calculate == "*":
    result = mul(num1, num2)
  elif calculate == "/":
    result = dev(num1, num2)
  else:
    print("그런 연산자는 없습니다.")

  return result
```

```python
calculator()
```

ex)

```python
# ctr + /  = 전체 주석 설정
# ctr + shift + L = 한꺼번에 같이 수정
```

```python
# HINT

# 연령대에 따라 카테고리를 지정해주는 함수 생성
def get_category(age):
  if age <= 5:
    category = 'Baby'

  elif age > 5 and age <= 12:
    category = 'Child'

  elif age > 12 and age <= 18:
    category = 'Teenager'

  elif age > 18 and age <= 25:
    category = 'Student'
  
  elif age > 25 and age <= 35: 
    category = 'Young Adult'
  
  elif age > 35 and age <= 60:
    category = 'Adult'

  else:
    category = 'Elderly'
  return category

# 타이타닉 데이터의 'Age'컬럼에 apply, lamda, 그리고 새로 생성한 get_category()를 사용하여 'Age_Category'라는 새로운 컬럼 생성
# 'Age_Category' 컬럼을 활용하여 각 카테고리 별로 몇명이 있는지 출력
```

```python
titanic_df['Age_category'] = titanic_df['Age'].apply(lambda x: get_category(x))
titanic_df
```

```python
print(titanic_df.groupby('Pclass').mean()['Fare'][1])
print(titanic_df.groupby('Pclass').mean()['Fare'][2])
print(titanic_df.groupby('Pclass').mean()['Fare'][3])
```

```python
def fare_category(Fare, Pclass):
  if Fare <= titanic_df.groupby('Pclass').mean()['Fare'][1] and Pclass == 1 :
    result = 'Cheap Ticket(1)'

  elif Fare > titanic_df.groupby('Pclass').mean()['Fare'][1] and Pclass == 1:
    result = 'Expensive Ticket(1)'

  elif Fare <= titanic_df.groupby('Pclass').mean()['Fare'][2] and Pclass == 2 :
    result = 'Cheap Ticket(2)'

  elif Fare > titanic_df.groupby('Pclass').mean()['Fare'][2] and Pclass == 2 :
    result = 'Expensive Ticket(2)'

  elif Fare <= titanic_df.groupby('Pclass').mean()['Fare'][3] and  Pclass == 3:
    result = 'Cheap Ticket(3)'

  elif Fare > titanic_df.groupby('Pclass').mean()['Fare'][3] and  Pclass == 3:
    result = 'Expensive Ticket(3)'
  
  return result
```

```python
titanic_df[['Fare', 'Pclass']]
```

```python
titanic_df['Ticket_fare'] = titanic_df.apply(lambda x : fare_category(x['Fare'], x['Pclass']), axis=1)
titanic_df
```
