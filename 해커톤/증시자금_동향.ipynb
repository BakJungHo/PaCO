## 증시자금동향 데이터 불러오기



```python
import pandas as pd
```

```python
money_df= pd.read_csv('/content/drive/MyDrive/Finance data/money_trend.csv',)
```

```python
money_df
```

## 데이터 내에 존재하는 comma 제거하기

```python
money_df['신용잔고'] = money_df['신용잔고'].apply(lambda x: x.replace(',',''))
money_df['고객예탁금'] = money_df['고객예탁금'].apply(lambda x: x.replace(',',''))
money_df['주식형 펀드'] = money_df['주식형 펀드'].apply(lambda x: x.replace(',',''))
money_df['혼합형 펀드'] = money_df['혼합형 펀드'].apply(lambda x: x.replace(',',''))
money_df['채권형 펀드'] = money_df['채권형 펀드'].apply(lambda x: x.replace(',',''))
```

```python
money_df
```

```python
money_df.info()
```

## 날짜 컬럼명 수정하고 오름차순으로 정렬하기

```python
money_df.rename(columns={'Unnamed: 0':'날짜'}, inplace=True)

money_df
```

```python
money_df.sort_values(['날짜'], ascending=True, ignore_index=True, inplace=True)
money_df
```

```python
money_df.dtypes
```

## str 데이터를 int 데이터로 변환하기

```python
money_df= money_df.astype({'신용잔고':'int'})
money_df= money_df.astype({'고객예탁금':'int'})
money_df= money_df.astype({'주식형 펀드':'int'})
money_df= money_df.astype({'혼합형 펀드':'int'})
money_df= money_df.astype({'채권형 펀드':'int'})

money_df
```

```python
money_df.dtypes
```

## str 날짜 데이터를 datetime 형식으로 변환하기

```python
from datetime import datetime
```

```python
for index in money_df['날짜'].index:
    dateStr = money_df['날짜'][index]
    date = datetime.strptime(dateStr, '%y-%m-%d')
    money_df['날짜'][index] = date.strftime('%Y-%m-%d')
```

```python
money_df['날짜']
```

```python
money_df['날짜'] = pd.to_datetime(money_df['날짜'])
```
